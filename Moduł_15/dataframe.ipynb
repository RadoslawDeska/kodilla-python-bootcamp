```python
import pandas as pd
```

```python
salaries_list = [
	("Mark", 1000, 23),
	("John", 1500, 25),
	("Daniel", 2300, 38),
	("Greg", 5000, 42)
]

salaries: pd.DataFrame = pd.DataFrame(salaries_list, columns=["name", "salary", "age"])
salaries.insert(len(salaries.columns), "initials", salaries["name"].str[0])
salaries = salaries.set_index("name")
salaries
```

```python
salaries['salary'].plot(kind='bar')
```

```python
salaries["age"].plot(kind='bar', color='orange', title='Workers age')
```
