```python
import pandas as pd
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]

prices = pd.DataFrame(prices, columns=["month", "price"])
prices = prices.set_index("month")
prices.insert(len(prices.columns), "priceUSD", prices["price"] / 4.0)

prices["priceUSD"].plot(kind='line', style='--r', grid=True, linewidth=3,
                        title='Price of goods (USD)',
                        ylim=(0,2))
```
